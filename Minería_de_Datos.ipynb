##### Mineria de Datos


La minería de datos es un proceso que utiliza herramientas analíticas avanzadas para extraer información útil de grandes volúmenes de datos almacenados en bases de datos. Este proceso permite identificar patrones, tendencias y relaciones que son clave para la toma de decisiones estratégicas en diversas áreas de negocio.

##### Ciencia de Datos

Es el campo integral que utiliza métodos científicos, algoritmos y sistemas para extraer conocimiento y insights de datos estructurados y no estructurados, con el objetivo de resolver problemas complejos, hacer predicciones y apoyar la toma de decisiones. Abarca todo el ciclo de vida de los datos, desde la formulación del problema y la recolección hasta el modelado avanzado, el despliegue de soluciones y la comunicación de resultados, combinando estadística, programación y conocimiento del dominio.
