##### Mineria de Datos


La minería de datos es un proceso que utiliza herramientas analíticas avanzadas para extraer información útil de grandes volúmenes de datos almacenados en bases de datos. Este proceso permite identificar patrones, tendencias y relaciones que son clave para la toma de decisiones estratégicas en diversas áreas de negocio.

##### Ciencia de Datos

Es el campo integral que utiliza métodos científicos, algoritmos y sistemas para extraer conocimiento y insights de datos estructurados y no estructurados, con el objetivo de resolver problemas complejos, hacer predicciones y apoyar la toma de decisiones. #### Estadística Básica

Media (Mean) parte central

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado) #peso distinto a cada elemento del arreglo
```

Mediana

```python
mediana = np.median(data)
```

Varianza y Desviación Estándar

De disperción

```python
varianza = np.var(data)
desviacion = np.std(data) # std - estandar
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000) #media, desviación y mil datos
print(np.mean(muestra), np.std(muestra))
```
